```python
import pandas as pd
import requests
import time

url = "https://api.binance.com/api/v3/ticker/price?symbol=BTCUSDT"

adatok = []

for i in range(100):
    r = requests.get(url)
    if r.status_code == 200:
        adat = r.json()
        adat["timestamp"] = pd.Timestamp.now()
        adatok.append(adat)
    time.sleep(1)

df = pd.DataFrame(adatok)

df.to_csv("stream100.csv", index=False)

df.head()
```
